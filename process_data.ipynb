```python
# process m4a to wav file
from pydub import AudioSegment
import os
PATH =  "Datasets/yn_raw_m3A"
PATH_To = "Datasets/yn_raw"
# new_length = 350000
All_Name_raw = os.listdir(PATH)
for name_class in All_Name_raw :
    name_in_class = os.listdir(PATH+"/"+str(name_class))
    for name in name_in_class :
        path_read = PATH+"/"+name_class + "/" + name
        output_file = PATH_To+"/"+name_class+"/"+name
        # main action start here
        audio = AudioSegment.from_file(path_read, format="m4a")
        audio.export(output_file[:-4]+".wav", format="wav")
        # resample_wav(path_read, output_file, new_length)
```

```python
PATH =  "Datasets/yn_raw"
PATH_To = "Datasets/yn"
new_length = 350000
All_Name_raw = os.listdir(PATH)
for name_class in All_Name_raw :
    name_in_class = os.listdir(PATH+"/"+str(name_class))
    for name in name_in_class :
        path_read = PATH+"/"+name_class + "/" + name
        output_file = PATH_To+"/"+name_class+"/"+name
        # main action start here
        resample_wav(path_read, output_file, new_length)
```

```python
#turn it into image
PATH =  "Datasets/yn/"
PATH_To = "Datasets/yn_image/"
new_length = 350000
All_Name_raw = os.listdir(PATH)
for name_class in All_Name_raw :
    name_in_class = os.listdir(PATH+"/"+str(name_class))
    for name in name_in_class :
        path_read = PATH+"/"+name_class + "/" + name
        output_file = PATH_To+"/"+name_class+"/"+name
        # main action start here
        process_wav_to_image(dir_wav = path_read, dir_tar = output_file[:-4]+".jpg",visual=True)
```
